# Create RSS feed from [clinicaltrials.gov](https://clinicaltrials.gov/)

```python
#RSS feed URL
rss_url = 'https://clinicaltrials.gov/api/rss?term=roche&dateField=StudyFirstPostDate'
```

```python
#install feedparser library
!pip install feedparser
```

```python
# Parse the RSS feed

import feedparser
feed = feedparser.parse(rss_url)
```

```python
feed
```

```python
# Print the feed title
print(f"Feed Title: {feed.feed.title}")

# Loop through each entry in the feed
for entry in feed.entries:
    print("\nTitle:", entry.title)
    print("Link:", entry.link)
    print("Published:", entry.published)
    print("Summary:", entry.summary)
```

# Convert into a dataframe

```python
import pandas as pd

# Extract relevant data and store it in a list of dictionaries
feed_data = []

for entry in feed.entries:
    feed_data.append({
        'Title': entry.title,
        'Link': entry.link,
        'Id': entry.id,
        'Published': entry.published,
        'Summary': entry.summary
    })
```

```python
# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(feed_data)
df
```

# Send via email

```python
import smtplib
import pandas as pd
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# Function to send email with HTML DataFrame
def send_email(dataframe, recipient_email, subject, sender_email, sender_password, smtp_server='smtp.gmail.com', smtp_port=587):
    # Convert DataFrame to HTML
    html_table = dataframe.to_html(index=False)  # Remove index for cleaner presentation

    # Create the email body (HTML)
    email_html = f"""
    <html>
    <head>
        <style>
            table {{
                width: 100%;
                border-collapse: collapse;
            }}
            th, td {{
                padding: 8px;
                text-align: left;
                border-bottom: 1px solid #ddd;
            }}
            th {{
                background-color: #f2f2f2;
            }}
        </style>
    </head>
    <body>
        <h2>This is today's RSS feed:</h2>
        {html_table}
    </body>
    </html>
    """

    # Set up the email message
    msg = MIMEMultipart('alternative')
    msg['Subject'] = subject
    msg['From'] = sender_email
    msg['To'] = recipient_email

    # Attach the HTML body to the email
    msg.attach(MIMEText(email_html, 'html'))

    # Send the email via SMTP server
    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()  # Start TLS encryption
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, recipient_email, msg.as_string())
        server.quit()
        print(f"Email sent to {recipient_email}!")
    except Exception as e:
        print(f"Error sending email: {e}")

# Email credentials and details
sender_email = 'sender_email'
sender_password = input("Type your password and press enter:")  # Use app-specific password if using Gmail with 2FA
recipient_email = 'recipient_email'
subject = 'Clinical Trials RSS Feed'

# Send the DataFrame via email
send_email(df, recipient_email, subject, sender_email, sender_password)
```
